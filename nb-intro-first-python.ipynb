We'll read a DNA sequence and ask it a few simple questions — no setup required.


```python
# A DNA sequence is just text made of A, T, G and C
dna = "ATGCGGTACCGATTACAGGCTAGCTAGCATCGTTAGC"
print("Sequence:", dna)
print("Length:", len(dna), "bases")
```

```python
# Count each base
for base in "ATGC":
    print(base, "appears", dna.count(base), "times")
```

```python
# GC content — the percentage of G and C bases
gc = (dna.count("G") + dna.count("C")) / len(dna) * 100
print(f"GC content: {gc:.1f}%")
```
