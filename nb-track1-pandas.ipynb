Tabular data without tears — build a small gene table and slice it.


```python
import pandas as pd

genes = pd.DataFrame({
    "gene":    ["INS", "HBB", "APP", "TP53", "BRCA1"],
    "chrom":   [11, 11, 21, 17, 17],
    "length_bp":[1431, 1606, 290000, 25772, 81189],
})
genes
```

```python
# Sort by length
genes.sort_values("length_bp", ascending=False)
```

```python
# Filter: genes on chromosome 17
genes[genes["chrom"] == 17]
```

```python
# A quick summary
print("Number of genes:", len(genes))
print("Mean length:", genes["length_bp"].mean())
genes["chrom"].value_counts()
```
