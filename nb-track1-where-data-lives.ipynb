```python
from Bio import Entrez, SeqIO
# NCBI asks you to identify yourself — put any email here
Entrez.email = "your.email@example.com"
```

```python
# Fetch the human HBB (beta-globin) mRNA record (RefSeq NM_000518)
handle = Entrez.efetch(db="nucleotide", id="NM_000518", rettype="gb", retmode="text")
record = SeqIO.read(handle, "genbank")
handle.close()

print("ID:", record.id)
print("Description:", record.description)
print("Length:", len(record.seq), "bp")
```

```python
# Peek at the annotated features
for feature in record.features[:8]:
    print(feature.type, feature.location)
```
