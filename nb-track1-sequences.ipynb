```python
# Write a tiny FASTA file
fasta = """>seq1 example gene\nATGCGTACGTAGCTAGCTAGCATCG\n>seq2 another\nTTTTGGGGCCCCAAAATTTTGGGG\n"""
open("example.fasta", "w").write(fasta)

from Bio import SeqIO
for rec in SeqIO.parse("example.fasta", "fasta"):
    print(rec.id, "| length", len(rec.seq), "|", rec.seq)
```

```python
# A FASTQ record adds a quality score for every base
fastq = """@read1\nATGCGTACGT\n+\nIIIIIHHHGG\n"""
open("example.fastq", "w").write(fastq)

for rec in SeqIO.parse("example.fastq", "fastq"):
    print(rec.id, rec.seq)
    print("Phred qualities:", rec.letter_annotations["phred_quality"])
```
