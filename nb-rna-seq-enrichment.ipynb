## 1. Set up and load a ranked gene list
Real human genes (named vector: Entrez ID -> fold change), shipped with the package.


```r
if (!requireNamespace("BiocManager", quietly = TRUE)) install.packages("BiocManager")
BiocManager::install(c("clusterProfiler","org.Hs.eg.db","DOSE","enrichplot"), update = FALSE, ask = FALSE)
library(clusterProfiler); library(org.Hs.eg.db); library(enrichplot)
data(geneList, package = "DOSE")
head(geneList)
```

## 2. Over-representation: GO biological processes


```r
sig_genes <- names(geneList)[abs(geneList) > 2]
ego <- enrichGO(gene = sig_genes, OrgDb = org.Hs.eg.db, ont = "BP",
                pAdjustMethod = "BH", pvalueCutoff = 0.05, readable = TRUE)
head(ego)
dotplot(ego, showCategory = 12)
```

## 3. GSEA: the whole ranked list


```r
gse <- gseGO(geneList = geneList, OrgDb = org.Hs.eg.db, ont = "BP",
             pvalueCutoff = 0.05, verbose = FALSE)
head(gse[, c("Description","NES","p.adjust")])
gseaplot2(gse, geneSetID = 1, title = gse$Description[1])
```

## 4. KEGG pathways (optional)


```r
# enrichKEGG / gseKEGG test pathways instead of GO terms
kk <- gseKEGG(geneList = geneList, organism = "hsa", pvalueCutoff = 0.05, verbose = FALSE)
head(kk[, c("Description","NES","p.adjust")])
```
