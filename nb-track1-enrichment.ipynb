Merge an expression table with an annotation table — the everyday bioinformatics 'join'.


```python
import pandas as pd

expression = pd.DataFrame({
    "gene": ["INS", "HBB", "APP", "TP53"],
    "tpm":  [120.5, 4.2, 33.1, 18.7],
})
annotation = pd.DataFrame({
    "gene": ["INS", "HBB", "APP", "TP53"],
    "description": ["insulin", "hemoglobin beta", "amyloid precursor", "tumor protein p53"],
})
```

```python
# Join the two tables on the shared 'gene' column
merged = expression.merge(annotation, on="gene")
merged
```

```python
# Now you can ask combined questions — e.g. the highest-expressed gene
merged.sort_values("tpm", ascending=False).head(1)
```
