## 2. Make a small example FASTQ
We generate 2,000 reads whose quality declines toward the 3' end, like real Illumina data. (On your own data, skip this and point FastQC at your real .fastq file.)


```python
import random
random.seed(0)
def qual(n):
    # Phred score starts high (~40) and declines toward the read end
    out = []
    for i in range(n):
        q = int(40 - i*0.22 + random.gauss(0, 2))
        q = max(2, min(40, q))
        out.append(chr(33 + q))   # Phred+33 encoding
    return ''.join(out)

with open('reads.fastq', 'w') as f:
    for i in range(2000):
        L = 100
        seq = ''.join(random.choice('ACGT') for _ in range(L))
        f.write(f'@read{i}\n{seq}\n+\n{qual(L)}\n')

!head -4 reads.fastq
```

## 3. Run FastQC


```python
!fastqc reads.fastq
!ls
```
