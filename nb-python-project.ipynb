Put the basics together: base composition, GC content, transcription, and the reverse complement.


```python
from collections import Counter
dna = "ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAGCTAGCTAGGCTA"
dna = dna.upper()
print("Length:", len(dna))
```

```python
# Base composition
counts = Counter(dna)
for base in "ATGC":
    print(f"{base}: {counts[base]}  ({counts[base]/len(dna)*100:.1f}%)")
```

```python
# GC content
gc = (counts["G"] + counts["C"]) / len(dna) * 100
print(f"GC content: {gc:.1f}%")
```

```python
# Transcription: DNA -> RNA (T becomes U)
rna = dna.replace("T", "U")
print("RNA:", rna)
```

```python
# Reverse complement
comp = {"A":"T", "T":"A", "G":"C", "C":"G"}
rev_comp = "".join(comp[b] for b in reversed(dna))
print("Reverse complement:", rev_comp)
```
